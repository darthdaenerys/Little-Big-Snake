```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('data.csv')
df.tail()
```

```python
avg_scores=[]
total_score=0
for idx,i in enumerate(df['Score']):
    total_score+=i
    avg_scores.append(total_score/(idx+1))
len(avg_scores)
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(18,7))
plt.plot(df['Score'],color='green',label='Game Score',linestyle='-',linewidth=2)
plt.plot(avg_scores,color='blue',label='Mean Score',linewidth=5)
plt.title('Score Trend')
plt.xlabel('Number of Games')
plt.ylabel('Score')
plt.legend()
plt.show()
```
